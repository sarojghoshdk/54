#### 1. Which Category has largest number of installations??

```python
df_cat_installs = df_copy.groupby(["Category"])["Installs"].sum().sort_values(ascending=False).reset_index()
df_cat_installs
```

```python
## Installs converting into billions
df_cat_installs.Installs = df_cat_installs.Installs/1000000000
```

```python
df2 = df_cat_installs.head(10)
df2
```

```python
## Visualization
plt.figure(figsize=(14,10))
sns.set_context("talk")
sns.set_style("darkgrid")

ax = sns.barplot(x="Installs", y="Category", data=df2)
ax.set_xlabel('No. of Installations in Billions')
ax.set_ylabel('')
ax.set_title("Most Popular Categories in Play Store", size = 20)
```

## Insights
- GAME Category has largest number of installations.

#### 2. What are the Top 5 most installed Apps in Each popular Categories ??


```python
df['Category'].unique()
```

```python
categories = [] 
for i in df['Category'].unique(): 
    if i != '1.9': 
        categories.append(i)
        
top_ten_category = categories[:10] 

for category in top_ten_category: 
    category_apps = df[df['Category'] == category] 
    category_apps_sorted = category_apps.sort_values(by='Installs', ascending=False)
    top_5_apps = category_apps_sorted.head(5)
    print(f"Top 5 apps in the {category} category:")
    for index, app in top_5_apps.iterrows(): 
        print(app['App'])
    print("=====================================================")
```

```python
df_cat_installs_app = df_copy.groupby(["Category","App"])["Installs"].sum().reset_index()
df_cat_installs_app = df_cat_installs_app.sort_values("Installs" , ascending=False)
df_cat_installs_app

df3 = df_cat_installs_app[df_cat_installs_app.Category == "GAME"]
df4 = df3.head(5)
df4
```

```python
df3 = df_cat_installs_app[df_cat_installs_app.Category == "COMMUNICATION"]
df4 = df3.head(5)
df4
```

```python
df3 = df_cat_installs_app[df_cat_installs_app.Category == "TOOLS"]
df4 = df3.head(5)
df4
```

```python
df_cat_installs_app = df_copy.groupby(["Category","App"])["Installs"].sum().reset_index()
df_cat_installs_app = df_cat_installs_app.sort_values("Installs" , ascending=False)
apps = ['GAME', 'COMMUNICATION', 'TOOLS',  'PRODUCTIVITY', 'SOCIAL' ]
sns.set_context("poster")
sns.set_style("darkgrid")

plt.figure(figsize=(40,40))

for i,app in enumerate(apps):
    df3 = df_cat_installs_app[df_cat_installs_app.Category == app]
    df4 = df3.head(5)
    plt.subplot(6,2,i+1)
    sns.barplot(data=df4, x= 'Installs' ,y='App')
    plt.xlabel('Installation in Millions')
    plt.ylabel('')
    plt.title(app,size = 40)
    
plt.tight_layout()
plt.show()
```

### Insights
- Most popular game is Subway Surfers.
- Most popular communication app is Hangouts.
- Most popular Tools app is Google.
- Most popular productivity app is Google Drive.
- Most popular social app is Instagram.
